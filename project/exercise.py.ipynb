```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
i = 1
keyword = input("검색할 영화 제목:")
url = "http://www.cgv.co.kr/search/movie.aspx?query={}&page={}".format(keyword, i)
```

```python
response = requests.get(url)
```

```python
result = bs(response.text, 'html.parser')
```
