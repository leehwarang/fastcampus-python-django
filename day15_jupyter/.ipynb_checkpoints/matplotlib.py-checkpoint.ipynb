```python
plt.plot([3,4,5,6,7,8], [13, 16, 21, 25, 29, 33]) #x축은 3,4,5,6,7,8 / y축은 13,16,21,25,29,33
```

```python
plt.plot([3,4,5,6,7,8], [13, 16, 21, 25, 29, 33], 'ro-') #x축은 3,4,5,6,7,8 / y축은 13,16,21,25,29,33
plt.axis([2.5,8.5, 10, 40]) #x축과 y축의 범위 지정하기
plt.xlabel('월')
plt.ylabel('온도')
```

```python
#y=x, y=x**2, y=x**3의 3가지 그래프 그리기
import numpy as np
```

```python
x_range = np.arange(0, 10, 0.2)
```

```python
x_range
```

```python
plt.plot(x_range, x_range, 'b-', x_range, x_range**2, 'yo', x_range, x_range**3, 'g--')
plt.xlabel('x축')
plt.ylabel('y축')
plt.title('여러 종류 그래프')
```

```python
# 선을 그리는 방법
# '색': r, b, y, g
# - 1개면 실선, -- 2개면 점섬
# ^세모, s 네모, o 동그라미
```
